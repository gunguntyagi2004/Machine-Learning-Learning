### min -max scaling
      X_new = (X - X_min) / (X_max - X_min)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df=pd.read_csv(r"C:\Users\tyagi\Downloads\wine_data.csv",header=None,usecols=[0,1,2])
df.columns=['Class Label','Alchol','Malic acid']
df.head(5)
```

```python
sns.kdeplot(df['Alchol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
color_dict={1:"red",2:"yellow",3:"purple"}
sns.scatterplot(x=df['Alchol'],y=df['Malic acid'],hue=df['Class Label'],palette=color_dict)
```

```python
x=df[['Alchol','Malic acid']]
y=df[['Class Label']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
```

```python
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(x_train_scaled.describe(),1)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.scatter(x_train['Alchol'],x_train['Malic acid'])
ax1.set_title("Before  Scaling")
ax2.scatter(x_train_scaled['Alchol'],x_train_scaled['Malic acid'],color='red')
ax2.set_title("After Scaling")
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
sns.kdeplot(x_train['Alchol'],ax=ax1)
sns.kdeplot(x_train['Malic acid'],ax=ax1)           
ax1.set_title("Before  Scaling")
sns.kdeplot(x_train_scaled['Alchol'],ax=ax2)
sns.kdeplot(x_train_scaled['Malic acid'],ax=ax2)  
plt.show()
```
