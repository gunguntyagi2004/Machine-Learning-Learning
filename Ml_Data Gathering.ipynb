```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36"
}

response =requests.get("https://www.ambitionbox.com/list-of-companies?page=2",headers=headers).text
response
```

```python
soup =BeautifulSoup(response,'lxml')
```

```python
company_name=[]
for  i in  soup.find_all('h2',class_="companyCardWrapper__companyName"):
    company_name.append(i.text.strip())

for i in company_name:
    print(i)
len(company_name)    
```

```python
rating=[]
for i in soup.find_all('div',class_="rating_text rating_text--md"):
    rating.append(i.text.strip())

for i in rating:
    print(i)
len(rating)    
```

```python
location=[]
for i in soup.find_all('span',class_="companyCardWrapper__interLinking"):
    city=i.text.strip().split("|")[1].split("+")[0]
    location.append(city)

for i in location:
    print(i)
len(location)    
```

```python
df=pd.DataFrame({'Company_name':company_name,'rating':rating,'location':location})
df
```
